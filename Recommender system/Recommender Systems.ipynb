# هر دو نوع توصیه گر در این فایل جوپیتر توضیح داده شده!

```python
import pandas as pd
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

کتابخانه های مورد نیاز همچون پانداس برای خواندن داده‌ها، جذر از کتابخانه مث برای گرفتن جذر، مت‌پلات لیب برای کشیدن نتایج و مت‌پلات لیب این لاین برای نشان داده پلات ها داخل خود نوت بوک فراخوانده شده‌اند

```python
movies_df = pd.read_csv('movies.csv') # خواندن فایل سی اس پی فیلم‌ها و ذخیره سازی داده ها در یک متغییر
ratings_df = pd.read_csv('ratings.csv') # خوانده فایل سی اس وی امتیازات و ذخیره سازی داده‌ها در یک متغییر

movies_df.head() # نمایش دیتا فریم فیلم‌ها 
```

در سلول پایین یک جا کد تغییر یاقته زیرا سال را از داده حذف نمی‌کرد

```python
# می‌خواهیم سال را در عنوان فیلم‌ها داخل دیتا فریم فیلم ها برداریم و در یک ستون جدید قرارشون بدیم
# عنوان هر یک از داده‌های را به صورت استرینگ انتخاب کرده و سال فیلم را از آن استخراج کرده و در نهایت در یک ستون جدید قرار میدهیم
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))', expand=False)
# داده‌های ستون سال را به صورت استرینگ جدا کرده و در نهایت قسمت اعداد را داخل ستون سال وارد می‌کنیم
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)', expand=False)
# سال را در هر یک عنوان داده‌ها حذف می‌کنیم
movies_df['title'] = movies_df.title.str.replace(r'(\(\d\d\d\d\))', '',regex=True)
# از ستون عنوان‌ها تک تک داده‌ها را گرفته و هر گونه فضای اضافی را حذف می‌کند
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
# چند سطر اول را نمایش داده‌ایم
movies_df.head()
```

```python
# ژانر هر فیلم را به که با | جدا شده است را جدا کرده و در داخل یک لیست قرار می‌دهیم و در نهایت داخل ستون ژانر قرارش می‌دهیم
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

حل می‌خواهیم برای هر ژانر یک ستون جدید بسازیم، اگر فیلم آن ژانر را داشت براش ۱ میزاریم و اگر نداشت ۰، زیرا به این شکل کار کردن با داده‌ها برایمان ساده تر می‌شود

```python
# دیتا فریم فیلم‌ها را داخل یک متغیر جدید کپی می‌کنیم
moviesWithGenres_df = movies_df.copy()

# از هر سطر دیتا فریم فیلم‌ها شاخص هر سطر به همراه اطلاعات هر ردیف را دریافت و داخل کی حلقه استفاده می‌کنیم
for index, row in movies_df.iterrows():
# از هز سطر لیست ژانر را داخل یک حلقه استفاده می‌کنیم
    for genre in row['genres']:
# داخل دیتا فریم جدیدی که ایجاد کردیم به ازای هر سطر و هر ژانر یک ستون جدید ایجاد کرده
        moviesWithGenres_df.at[index, genre] = 1

# بعضی از فیلم‌ها بعضی ار ژانرها را ندارند بنابراین به جای آن ژانر نان گذاشته شده که آن را به صفر تبدیل می‌کنیم
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
# دیتا فریم جدید را نمایش می‌دهیم
moviesWithGenres_df.head()
```

```python
# چند سطر دیتا فریم امتیازات را نمایش میدهیم|
ratings_df.head()
```

```python
# ستون زمان‌ها را از دیتا فریم امتیازات حذف می‌کنیم
ratings_df = ratings_df.drop('timestamp', axis=1)
ratings_df.head()
```

# سیستم توصیه‌گر مبتنی بر محتوا

میخواهیم یک کاربر ایجاد کنیم که به یک سری فیلم امتیاز داده است و میخواهیم با استفاده از این داده‌ها یک سری فیلم به این کاربر پیشنهاد بدیم

```python
# یک لیست از چند دیشنری درست می‌کنیم که شامل عنوان فیلم و امتیازی که به فیلم داده شده است میباشد
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ]
# لیست کاربر را تبدیل به یک دیتا فریم میکنیم و آن را نمایش میدهیم
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
# لیست عنوان‌های کاربر را با کل فیلم‌‌های موجود مقایسه کرده و آن‌های که در دیتا فریم فیلم‌ها وجود دارند را به عنوان یک دیتا فریم جدید داخل یک متغییر میریزیم
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
# تمامی ستون‌های دو دیتا ست را با یکدیگر در یک دیتافریم قرار میدهد 
inputMovies = pd.merge(inputId, inputMovies)
# ستون سال و ژانر را به طور کلی از دیتا فریم بر میداریم
inputMovies = inputMovies.drop('genres', axis=1).drop('year', axis=1)

inputMovies
```

```python
# لیست فیلم‌‌هایی که تماشاگر امتیاز داده است با دیتای کلیه فیلم‌ها مقایسه کرده و هر فیلم‌ را همراه با ژانر‌های آن‌ها و شاخصشون بر می‌گرداند
userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]
# دیتا فریم جدید را نشان می‌دهد
userMovies
```

```python
# شاخص دیتا فریم جدید را از نو شماره گذاری می‌کند
userMovies = userMovies.reset_index(drop=True)
# از دیتا فریم جدید تنها ژانر آنها را نگه میدارد و بقیه ستون‌ها را حذف می‌کند
userGenreTable = userMovies.drop('movieId', axis=1).drop('title', axis=1).drop('genres', axis=1).drop('year', axis=1)
# دیتا فریم جدید را نمایش می‌دهد
userGenreTable
```

```python
# از دیتا فریم مربوط به تماشاگر تنها امتیاز فیلم‌هایی که دیده است را نمایش می‌دهد
inputMovies['rating']
```

```python
# دیتا فریم ژانرها را معکوس کرده و امتیاز هر فلیم را در ژانر آن ضرب کره و جمع متیاز تمامی فیلم‌ها برای هر ژانر جمع شده
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])
# لیست امتیازات هر ژانر را برای این تماشاگر نمایش می‌دهد
userProfile
```

```python
# دوباره یک دیتا فریم دیگر از فیلم‌ه ایجاد کرده که اینکار با مقایسه شماره شناسایی هر فیلم با کلیه فیلم‌ها انجام می‌شود
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])
# سال، ژانر، عنوان و شماره شناسایی هر فیلم برداشته می‌شود
genreTable = genreTable.drop('movieId', axis=1).drop('title', axis=1).drop('genres', axis=1).drop('year', axis=1)
# نمایش دیتا فریم جدید
genreTable.head()
```

```python
# اندازه سطر و ستون دیتا فریم چدید را نمایش می‌دهد
genreTable.shape
```

```python
# دیتا فریم ژانر‌ها را در امتیاز هر ژانر (توسط بیننده) ضرب کرده بنابراین هر جدولی از ژانرها تشکیل شده که هرچه هر ژانری مورد سلیقه تماشگر بوده
# ی امتیاز بیشتری کسب کرده و آنهایی که نبوده امتیاز صفر گرفته، سپس برای هر فیلم ستون ژانر‌ها با یکدیگر جمع شده و بر جمع امتیازات ژانر‌های پروفایل بیننده تقسیم می‌شود که به گونه‌‌ای آن را نرمالایز کرده
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
# لیست امتیازات تمامی فیلم‌ها بر اساس غلایق تماشاگر نشان داده شده
recommendationTable_df.head()
```

```python
# لیست امتیازات تمامی فیلم‌ها بر اساس علایق بیننده به صورت نزولی مرتب می‌شود
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
# چند عدد از فیلم‌های که بسیار به علایق بیننده است نمایش داده شده
recommendationTable_df.head()
```

```python
# شاخص ۲۰ عدد از فیلم‌های تشخیص داده شده با شماره شناسایی دیتا فریم فیلم‌ها مقایسه شده تا آن‌های که باید انتخاب شوند سپس بر اساس شماره شناسایی فیلم‌ مورد نظر مکانش پیدا و نمایش داده می‌شود
movies_df.loc[movies_df['movieId'].isin(recommendationTable_df.head(20).keys())]
```

# سیستم توصیه‌گر مبتنی بر همکاری

میخواهیم یک کاربر ایجاد کنیم که به یک سری فیلم امتیاز داده است و میخواهیم با استفاده از این داده‌ها یک سری فیلم به این کاربر پیشنهاد بدیم

```python
# یک لیست از چند دیشنری درست می‌کنیم که شامل عنوان فیلم و امتیازی که به فیلم داده شده است میباشد
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ]
# لیست کاربر را تبدیل به یک دیتا فریم میکنیم و آن را نمایش میدهیم
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
# دیتا فریمی جدید با چک کردن فیلم‌های دیده شده توسط کاربر در بین‌تمامی فیلم‌ها با استفاده از عنوان آن‌ها تشکیل می‌دهد
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
# دیتا فریمی جدیدی با استفاده از دیتا فریم جدید و دیتا فریم پروفایل کاربر ایجاد می‌کند
inputMovies = pd.merge(inputId, inputMovies)
# ستون سال را دیتا فریم حذف می‌کند
inputMovies = inputMovies.drop('year', axis=1)
# دیتا فریم جدید را نمایش می‌دهد
inputMovies
```

```python
# فیلم‌های دیده شده توسط کاربر را از دیتا فریم امتیازات جدا کرده و نمایش می‌دهد
userSubset = ratings_df[ratings_df['movieId'].isin(inputMovies['movieId'].tolist())]
```

```python
# یک دیتافریم جدید درست میکنه که بر اساس شماره شناسایی کاربران گروه‌بندی شده
userSubsetGroup = userSubset.groupby(['userId'])
```

```python
# از دیتا فریم جدید گروه ۱ را صدا میزنه و نشان میدهد
userSubsetGroup.get_group((1,))
```

```python
# دیتا فریم زیرگروه کاربر را به صورت نزولی بر اساس تعداد فیلم‌ها برای هر کاربر مرتب می‌کند و اگر هر کاربری تعداد فیلم‌ بیشتری با کاربر مورد نظر داشته باشد در صدر قرار می‌گیرد
userSubsetGroup = sorted(userSubsetGroup,  key=lambda x: len(x[1]), reverse=True)
```

```python
# تعدادی از کاربران را که بیشترین تعداد فیلم را با کاربر مورد نظر به اشتراک دارند را نمایش می‌دهد
userSubsetGroup[0:3]
```

```python
# تعداد ۱۰۰ کاربر اول را انتخاب می‌کند برای توصیه کردن به کاربر مورد نظر
userSubsetGroup = userSubsetGroup[0:100]
```

می‌خواهیم ضریب همبستگی پیرسون را بین کاربر مورد نظر و زیر گروه کاربران بیابیم و آن را داخل یک دیکشنری ذخیره سازی کنیم

```python
#Store the Pearson Correlation in a dictionary, where the key is the user Id and the value is the coefficient
# برای ذخیره سازی همبستگی پیرسون یک دیکشنری درست می‌کنیم که کلید آن شماره شناسایی کاربر و مقدار آن مقدار همبستگی
pearsonCorrelationDict = {}

# برای هر کاربر در زیر گروع یک حلقه ایجاد کردیم
for name, group in userSubsetGroup:
    # اسم به صورت تاپل هست که اون را به اینتیجر تبدیل می‌کنیم
    name = list(name)
    name = name[0]
    # گروهی که گرفته را بر اساس سماره سناسایی فیلم به صورت صعودی مرتب می‌کنه 
    group = group.sort_values(by='movieId')
    # فیلم‌های کاربر را به صورت صعودی مرتب می‌کند
    inputMovies = inputMovies.sort_values(by='movieId')
    # تعداد فیلم‌هایی که هر گروه دیده را در یک متغییر جدید ذخیره می‌کند
    nRatings = len(group)
    # برای هر گروه فیلم‌هایی که کاربران دیگر با کاربر مورد نظر با یکدیگر به اشتراک دارند را در یک متغییر جدید ذخیره می‌کند
    temp_df = inputMovies[inputMovies['movieId'].isin(group['movieId'].tolist())]
    #And then store them in a temporary buffer variable in a list format to facilitate future calculations
    # امتیازاتی که به فیلم کاربر مورد نظر به فیلم‌ها داده است را به به صورت لیست ذخیره می‌کند
    tempRatingList = temp_df['rating'].tolist()
    # امتیازاتی که هر گروه را به صورت لیست ذخیره سازی می‌کند
    tempGroupList = group['rating'].tolist()
    # محاسبات مربوط به ضریب همبستگی پیرسون محاسبه خواهد شد
    Sxx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList),2)/float(nRatings)
    Syy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList)*sum(tempGroupList)/float(nRatings)
    
    #If the denominator is different than zero, then divide, else, 0 correlation.
    #برای هر شماره شناسایی کاربر چک میکنه اگر مخره صفر شده ضریب همبستگی پیرسون را صفر میگذاره و در غیر اینصورت عدد صریب را قرار می‌دهد
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[name] = Sxy/sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[name] = 0
```

```python
# برای صد کاربر ضریب همبستگی پیرسون را نمایش می‌دهد
pearsonCorrelationDict.items()
```

```python
# دیکشنری را تبدیل به یک دیتافریم کرده به گونه‌ای که اسم هر ردیف کلید‌های دیکشنری باشد 
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
# اسم ستون را شاخص شباهت نام دادیم
pearsonDF.columns = ['similarityIndex']
# یک ستون جدید ایجاد کردیم و شماره شناسایی کاربر را در آن قرار دادیم
pearsonDF['userId'] = pearsonDF.index
# شاخص دیتا فریم را از صفر نام گذاری کردیم
pearsonDF.index = range(len(pearsonDF))
# دیتا فریم جدید را نمایش می‌دهیم
pearsonDF.head()
```

```python
# کاربران را بر اساس  شباهتشون به صورت نزولی مرتب می‌کنه
topUsers=pearsonDF.sort_values(by='similarityIndex', ascending=False)[0:50]
# چند کاربر اول را که بسیار به کاربر مورد نظر شباهت دارند را نمایش می‌دهد
topUsers.head()
```

```python
# دو دیتا فریم را با یکدیگر ادغام می‌کند به گونه‌ای که شماره شناسایی‌های هر دو دیتا فریم هر دو طرف وجود داشته باشد
topUsersRating = topUsers.merge(ratings_df, left_on='userId', right_on='userId', how='inner')
topUsersRating.head()
```

```python
# یک ستون جدید به نام وزن امتیازات با استفاده از ضرب ضریب پیرسون در امتیازات کاربران ایجاد می‌کنم
topUsersRating['weightedRating'] = topUsersRating['similarityIndex']*topUsersRating['rating']
topUsersRating.head()
```

```python
a = topUsersRating.groupby('movieId').sum()[['similarityIndex','weightedRating']]
a.head()
```

```python
# دیتافریم نزدیک‌ترین کاربران را بر اساس فیلم‌های دیده شده توسط کاربر مورد نظر گروه بندی کرده و سطر‌های هر گروه را با یکدیگر جمع می‌کند و تنها دو ستون را داخل متغییر جدید میریزد
tempTopUsersRating = topUsersRating.groupby('movieId').sum()[['similarityIndex','weightedRating']]
# دو ستون جدید را با اسم‌های جدید نام گذاری می‌کند
tempTopUsersRating.columns = ['sum_similarityIndex','sum_weightedRating']
tempTopUsersRating.head()
```

```python
# یک دیتا فریم جدید و خالی ایجاد می کنیم
recommendation_df = pd.DataFrame()
# اولین ستون دیتافریم جدید را ایجاد کرده که داده‌های آن جمع وزن امتیاز هر گروه فیلم تقسیم بر شاخص شباهت هر فیلم است
recommendation_df['weighted average recommendation score'] = tempTopUsersRating['sum_weightedRating']/tempTopUsersRating['sum_similarityIndex']
# یک ستون جدید ایجاد کرده که شماره شناسایی متناسب هر فیلم در آن قرار دارد
recommendation_df['movieId'] = tempTopUsersRating.index
# دیتافریم جدید را نشان داده است
recommendation_df.head()
```

```python
# دیتا فریم مورد نظر را بر اساس ستون اول دیتافریم به صورت نزولی مرتب می‌کند
recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
# چند عدد از فیلم‌های دیتافریم توصیه‌گر را نمایش می‌دهد
recommendation_df.head(10)
```

```python
# از داخل دیتافریم کل فیلم‌ها آن‌هایی که داخل دیتافریم توصیه‌گر هستند را می‌یابد و نمایش می‌هد که ۱۰ عدد از فیلم‌ها اینجا انتخاب شده تا نمایش داده شود
# ابتدا شماره شناسایی ۱۰ عدد از فیلم‌های داخل دیتافریم توصیه‌گر پیدا شده و سپس تبدیل به لیست می‌شود و در نهایت آنهایی که داخل دیتا فریم کل فیلم‌ها هستند مکان‌یابی شده نو نمایش داده می‌شود
movies_df.loc[movies_df['movieId'].isin(recommendation_df.head(10)['movieId'].tolist())]
```
